```python
import os

# Get unique BU and schedule combinations
bu_schedules = df[['BU', 'schedule']].drop_duplicates().reset_index(drop=True)

# Create a list to store possible file name matches
bu_schedules['file_name_matches'] = ''

# Iterate over each BU and schedule combination
for index, row in bu_schedules.iterrows():
    # Get the unique BU and schedule values
    bu = row['BU']
    schedule = str(row['schedule'])

    # Find files in the /payroll/sal_sched folder that match the BU and schedule
    matching_files_070824 = [file for file in os.listdir('payroll/sal_sched') 
                             if '070824' in file and bu.lower() in file.lower() and schedule.lower() in file.lower()]
    matching_files_0724 = [file for file in os.listdir('payroll/sal_sched') 
                           if '0724' in file and bu.lower() in file.lower() and schedule.lower() in file.lower()]

    # Prioritize 070824 files
    if matching_files_070824:
        bu_schedules.loc[index, 'file_name_matches'] = ', '.join(matching_files_070824)
    else:
        bu_schedules.loc[index, 'file_name_matches'] = ', '.join(matching_files_0724)


bu_schedules = bu_schedules[bu_schedules['file_name_matches'].notna() & (bu_schedules['file_name_matches'] != '')]

# Display the table of BU and schedule combinations with possible file name matches
bu_schedules.head()
```

```python
import camelot
import pandas as pd

# Create a new DataFrame to store the results
results = []

# Iterate through the files and extract table shapes
for index, row in bu_schedules.iterrows():
    pdf_path = f"payroll/sal_sched/{row['file_name_matches'].strip()}"
    tables = camelot.read_pdf(
        pdf_path,
        pages="0",
        flavor="stream",
        edge_tol=40
    )
    
    # Extract and store the shape of each table in the PDF
    for i, table in enumerate(tables):
        results.append({
            'BU': row['BU'],
            'schedule': row['schedule'],
            'file_name': row['file_name_matches'].strip(),
            'table_number': i+1,
            'shape': table.df.shape
        })

# Create a new DataFrame with the results
results_df = pd.DataFrame(results)

# Order by bargaining unit and salary schedule
results_df = results_df.sort_values(by=['BU', 'schedule'])

print(results_df)
```

```python
import camelot
import pandas as pd

def extract_and_clean_pdf_table(
    pdf_path: str,
    pages: str = "1",
    flavor: str = "stream",
    edge_tol: int = 40,
    skip_rows: int = 2,
    skip_footer_rows: int = 0,
    footer_cols_to_drop: int = 1,
) -> pd.DataFrame:
    """
    1) Reads the first table from the given PDF.
    2) Drops `skip_rows` rows at top (e.g. title/header junk).
    3) Drops `skip_footer_rows` rows at bottom (e.g. total rows, notes).
    4) Drops the last `footer_cols_to_drop` columns (e.g. totals or notes).
    5) Promotes the first remaining row to column headers.
    6) Renames the first column "range" and coerces to Int64, dropping blanks.
    """
    tables = camelot.read_pdf(
        pdf_path,
        pages=pages,
        flavor=flavor,
        edge_tol=edge_tol
    )
    if not tables:
        raise ValueError(f"No tables found in {pdf_path!r}")
    df = tables[0].df.copy()

    # 2) drop header rows
    df = df.iloc[skip_rows:]
    # 3) drop footer rows
    if skip_footer_rows:
        df = df.iloc[:-skip_footer_rows]
    # 4) drop footer columns
    if footer_cols_to_drop:
        df = df.iloc[:, :-footer_cols_to_drop]

    # 5) promote first row to header
    df.columns = df.iloc[0]
    df = df.iloc[1:].reset_index(drop=True)

    # 6) rename first column to "range"
    first_col = df.columns[0]
    df = df.rename(columns={first_col: "range"})

    # drop rows where range is blank or nan
    df = df[~df["range"].astype(str).str.strip().isin(["", "nan"])]
    df["range"] = pd.to_numeric(df["range"], errors="coerce").astype("Int64")

    return df


def transform_salary_df(
    df: pd.DataFrame,
    BU: str,
    schedule: str,
    id_var: str = "range",
    var_name: str = "step",
    value_name: str = "biweekly",
    annual_periods: int = 26,
) -> pd.DataFrame:
    """
    1) Unpivots the cleaned salary table on `id_var`.
    2) Strips commas and converts the `value_name` to float.
    3) Calculates an annual salary by multiplying by `annual_periods`.
    4) Prepends BU and schedule columns.
    """
    df_unpivoted = pd.melt(
        df,
        id_vars=id_var,
        var_name=var_name,
        value_name=value_name
    )
    # clean up numbers
    df_unpivoted[value_name] = (
        df_unpivoted[value_name]
        .str.replace(",", "")
        .astype(float)
    )
    df_unpivoted["ak_annual_salary"] = (
        df_unpivoted[value_name] * annual_periods
    )

    # insert metadata
    df_unpivoted.insert(0, "BU", BU)
    df_unpivoted.insert(1, "schedule", schedule)

    return df_unpivoted


def process_salary_schedule(
    pdf_path: str,
    BU: str,
    schedule: str,
    **extract_kwargs
) -> pd.DataFrame:
    """
    Thin wrapper: runs extract_and_clean_pdf_table() with any overrides,
    then transform_salary_df() with the BU and schedule you pass.
    """
    cleaned = extract_and_clean_pdf_table(pdf_path, **extract_kwargs)
    return transform_salary_df(cleaned, BU, schedule)
```

```python
import pandas as pd

# 1) your existing functions must already be defined:
#    - extract_and_clean_pdf_table(...)
#    - transform_salary_df(...)
#    - process_salary_schedule(...)

# 2) define a config dict for all the PDFs you want to pull
schedules_config = {
    # XE schedules
    "XE_A00": {
        "pdf_path": "payroll/sal_sched/070824xew_A00.pdf",
        "BU": "XE",
        "schedule": "A00",
        "skip_rows": 1,
        "skip_footer_rows": 1,
        "footer_cols_to_drop": 1,
    },
    "XE_N00": {
        "pdf_path": "payroll/sal_sched/070824xew_N00.pdf",
        "BU": "XE",
        "schedule": "N00",
        "skip_rows": 1,
        "skip_footer_rows": 1,
        "footer_cols_to_drop": 1,
    },

    # SS schedules
    "SS_100": {
        "pdf_path": "payroll/sal_sched/070824ssw_100.pdf",
        "BU": "SS",
        "schedule": "100",
        "skip_rows": 1,
        "footer_cols_to_drop": 1,
    },
    "SS_200": {
        "pdf_path": "payroll/sal_sched/070824ssw_200.pdf",
        "BU": "SS",
        "schedule": "200",
        "skip_rows": 2,
        "footer_cols_to_drop": 1,
    },
    "SS_500": {
        "pdf_path": "payroll/sal_sched/070824ssw_500.pdf",
        "BU": "SS",
        "schedule": "500",
        "skip_rows": 2,
        "footer_cols_to_drop": 1,
    },
    "SS_600": {
        "pdf_path": "payroll/sal_sched/070824ssw_600.pdf",
        "BU": "SS",
        "schedule": "600",
        "skip_rows": 1,
        "footer_cols_to_drop": 1,
    },
    "SS_C00": {
        "pdf_path": "payroll/sal_sched/070824ssw_C00.pdf",
        "BU": "SS",
        "schedule": "C00",
        "skip_rows": 2,
        "footer_cols_to_drop": 1,
    },

    # GP schedules
    "GP_100": {
        "pdf_path": "payroll/sal_sched/0724gpw_100.pdf",
        "BU": "GP",
        "schedule": "100",
        "skip_rows": 3,
        "skip_footer_rows": 2,
    },
    "GP_200": {
        "pdf_path": "payroll/sal_sched/0724gpw_200.pdf",
        "BU": "GP",
        "schedule": "200",
        "skip_rows": 2,
        "skip_footer_rows": 1,
    },
}

# 3) a helper to run them all and concat
def process_multiple_schedules(configs: dict) -> pd.DataFrame:
    """
    Given a dict of name -> kwargs for process_salary_schedule,
    runs each, and returns one big DataFrame.
    """
    all_dfs = []
    for name, params in configs.items():
        try:
            df = process_salary_schedule(**params)
            all_dfs.append(df)
        except Exception as e:
            print(f"⚠️  Skipped {name}: {e}")
    if not all_dfs:
        raise RuntimeError("No schedules processed successfully.")
    return pd.concat(all_dfs, ignore_index=True)

# 4) run it
all_schedules_df = process_multiple_schedules(schedules_config)

# 5) inspect
print(all_schedules_df.shape)
all_schedules_df.head()
```
